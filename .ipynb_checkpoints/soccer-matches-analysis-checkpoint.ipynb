So here we need some domain knowledge in soccer. If you frequently watch soccer games, you'll know that whether or not a team was playing at home would probably make a significant difference in its performance. Therefore, we want to take which team is playing at home into account. 

- We add two new variables to indicate whether or not the teams playing at home.
    - ```team1Home``` 
    - ```team2Home```
- If the first team is playing at home, then ```team1Home``` is ```True```, ```team2Home``` is ```False``` and vice versa.
- If both teams are playing at home, both ```team1Home``` and ```team2Home``` are ```True```; if both teams are away, then ```team1Home``` and ```team2Home``` are both ```False```. 

````python
matches['team1Home'] = matches.apply(lambda x: str(x['team1Text']) in str(x['venue']), axis=1)
matches['team2Home'] = matches.apply(lambda x: str(x['team2Text']) in str(x['venue']), axis=1)
matches.head()
````

Also, another factor that might influence a team's performance is the type of matches it's playing. Normally, teams will treat Friendly match as an opportunity for exploring young talented players. Therefore, they would be less serious towards the scores. For the actual prediction, we might want to exclude the friendly matches if they turn out to make a huge difference. 

````python
import copy
# make a deep copy for the matches including friendly matches 
matches_with_friendly = copy.deepcopy(matches)
# exclude the friendly matches for our analysis
matches = matches[matches['CupName'] != 'Friendly']
# also, we want to exclude the possible influence of outliers, like some
# teams would score more than 10 goals in a single match
matches = matches[(matches['team1Score']<=10) & (matches['team2Score']<=10)]
matches.shape
````

It turns out Friendly matches do occupy a significant portion of the total matches. Also, the data set needs to be further modified since a win for team 1 automatically means a loss for team 2. However, looking at an entry, it only reflects the result for a single team. Therefore, we should mirror each entry in another team's perspective, which means we could make a reverse version for every entry.

for example,
* team1 vs team2 
    * score - 1:3, team1Home - False, team2Home - True.
* team2 vs team1
    * score - 3:1, team1Home - True, team2Home - False

````python
matches_mirror = copy.deepcopy(matches)

matches_mirror['team1'] = matches['team2']
matches_mirror['team2'] = matches['team1']

matches_mirror['team1Text'] = matches['team2Text']
matches_mirror['team2Text'] = matches['team1Text']

matches_mirror['team1Score'] = matches['team2Score']
matches_mirror['team2Score'] = matches['team1Score']

matches_mirror['team1Home'] = matches['team1Home']
matches_mirror['team2Home'] = matches['team2Home']

# concatenate the mirrored matches into the original matches data set
# to double up the number of entries
matches = pd.concat([matches, matches_mirror])
matches.shape
````

### Winning Percentage 
- we can observe from the time series plot that the average total scores per game has been dropping till 1990. After that, we see a solid increase. This makes sense, because now teams put more effort in defense rather than attack, leading to rarer scoring opportunities. 
- next, we want to look at the winning percentage for each team, with and without friendly matches.

````python
matches.index = range(len(matches))

def win_pct(matches):
    ### create a dictionary to store the total number
    ### of games and number of winnings for each team
    win_pct_teams = dict()
    
    for i in matches.index:
        team1 = matches.loc[i, 'team1']
        team2 = matches.loc[i, 'team2']
        score1 = matches.loc[i, 'team1Score']
        score2 = matches.loc[i, 'team2Score']
        
        ### for every entry, we have information about 
        ### two teams. 
        ### Update information for team 1
        if team1 in win_pct_teams:
            win_pct_teams[team1]['total_matches'] += 1
            if score1 > score2:
                win_pct_teams[team1]['win_matches'] += 1
        else:
            if score1 > score2:
                win_pct_teams[team1] = {'total_matches':1, 'win_matches':1}
            else:
                win_pct_teams[team1] = {'total_matches':1, 'win_matches':0}
                
        ### update information for team 2
        if team2 in win_pct_teams:
            win_pct_teams[team2]['total_matches'] += 1
            if score2 > score1:
                win_pct_teams[team2]['win_matches'] += 1
        else:
            if score2 > score1:
                win_pct_teams[team2] = {'total_matches':1, 'win_matches':1} 
            else:
                win_pct_teams[team2] = {'total_matches':1, 'win_matches':0}
                
    win_pct_info = list()
                
    ### convert dictionary into a list 
    for k,v in win_pct_teams.items():
        win_pct_info.append((k, v['total_matches'], v['win_matches'], 
                             v['win_matches'] / v['total_matches']))
        
    ### convert dictionary into a dataframe
    import pandas as pd
    win_df = pd.DataFrame(win_pct_info, columns=['team_name', 'total_matches', 
                                                 'win_matches', 'win_pct'])
    
    return win_df[win_df['total_matches'] >= 10]
        
````

````python
### no friendly matches 
df_no_friendly = win_pct(matches)
### with friendly matches
df_friendly = win_pct(matches_with_friendly)
````

#### visualize winning percentage 

````python
def visualize_win_pct(df, title):
    x = df['total_matches']
    y = df['win_pct']
    n = df['team_name']
        
    import matplotlib.pyplot as plt
    %matplotlib inline 
    fig, ax = plt.subplots()
    fig.set_size_inches(16, 8)
    
    ax = sns.scatterplot(x='total_matches', y='win_pct', data=df)
    
    def label_pts(x, y, val, ax):
        a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
        for i, point in a.iterrows():
            ax.text(point['x'], point['y'], str(point['val']))

    label_pts(x, y, n, ax)
    
    ### draw a horizontal line to indicate 50% winning percentage 
    plt.axhline(y=0.5, xmin=0, xmax=1000, linewidth=2, color='y')
    
    ### title, x and y axis labels
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('total matches', fontsize=13)
    ax.set_ylabel('win pct', fontsize=13)
````

````python
### we want to look at the winning rates of recent games 
### for example, the winning rate of the previous 5, 10, 20 and 40 matches

def gen_recent_info(matches, team_name, team_idx, num_matches):
    team_matches = matches[matches[team_idx] == team_name]
    team_matches = team_matches.sort_values(by='date', ascending=False)
    
    ### number of wins, losses and draws
    win, draw, loss = 0, 0, 0
    
    if team_matches.shape[0] < num_matches:
        return np.nan, np.nan, np.nan
    
    for i in range(num_matches):
                
        row = team_matches.iloc[i]
        if row['team1Score'] > row['team2Score']:
            win += 1
        if row['team1Score'] < row['team2Score']:
            loss += 1
        if row['team1Score'] == row['team2Score']:
            draw += 1
            
    return win / num_matches, draw / num_matches, loss / num_matches
````

````python
### all possible teams 
team_names = list(set(matches['team1']))

### recent number of matches 
recent_matches = [5, 10, 15, 20, 30, 40, 50, 100]

### convert number of matches into names 
recent_info_names = np.array([[f'win_recent_{i}', 
                               f'draw_recent_{i}', 
                               f'loss_recent_{i}'] 
                              for i in recent_matches])

recent_info_names = np.append(recent_info_names, 'team_name')


### flatten 2-D list into a 1-D array
recent_info_names = recent_info_names.flatten()
recent_matches_df = pd.DataFrame(columns = recent_info_names)

### add recent matches info into each team 
for tn in team_names:
    rates = list()
    for rm in recent_matches:
        win, draw, loss = gen_recent_info(matches, tn, 'team1', rm)
        rates.append(win)
        rates.append(draw)
        rates.append(loss)
    rates.append(tn)
    dict_ = dict(zip(recent_info_names, rates))
    recent_matches_df = recent_matches_df.append(dict_, ignore_index=True)
````

### Build a Random Forest Classifier 
- build a random forest classifier
- predict categorical variable
- the target variable is ```result```
    - ```draw``` = 0
    - ```win``` = 1
    - ```lose``` = 2
- use 52 features extracted before
- Baseline model

````python
matches.loc[matches['goal_diff'] == 0, 'result'] = 0
matches.loc[matches['goal_diff'] > 0, 'result'] = 1
matches.loc[matches['goal_diff'] < 0, 'result'] = 2
matches = matches.astype({'result':int})
````

````python
matches.head()
````

````python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(matches[features], matches['result'], 
                                                    test_size=0.2, random_state=99)

parameters = {'n_estimators':[50, 100]}
rfc = RandomForestClassifier(random_state=99)
model = GridSearchCV(rfc, parameters, cv=6)
model.fit(X_train, y_train)
print('The best number of trees is: ', model.best_params_['n_estimators']) 

# make predictions
y_predict = model.best_estimator_.predict(X_test)
print('The sqrt of mean squared error on test set:', np.sqrt(mean_squared_error(y_predict,y_test))) 
print('The mean absolute error on test set:', mean_absolute_error(y_predict,y_test)) 
````

````python
model.score(X_test, y_test)
````

#### Convert goal difference to win-draw-lose

````python
def convert_to_win_draw_lose(goal_diff):
    result = []
    for gf in goal_diff:
        if -0.1 <= gf <= 0.1:
            result.append('draw')
        elif gf > 0.1:
            result.append('win')
        else:
            result.append('lose')
            
    return result

def cat_acc(act, predict):
    total = len(act)
    correct = 0
    
    for i in range(total):
        if predict[i] == act[i]:
            correct += 1
    
    return correct / total
````

````python
y_test_wdl = convert_to_win_draw_lose(y_test)
y_pred_wdl = convert_to_win_draw_lose(y_predict)

print(f'win-draw-lose acc: {cat_acc(y_test_wdl, y_pred_wdl)}')
````

__During this early group stage, if a team wins a match, it gets 3 points; if it gets a draw, it gains 1 point; if it loses, 0 point awarded__

````python
### simulate the group stage result and return
### the first 2 teams for each group

def simulate_group_stage(wc_matches):
    
    ### keep track of all the points for each team
    wc_groups = {
        'A': {'Russia': 0, 'Saudi Arabia': 0, 'Egypt': 0, 'Uruguay': 0},
        'B': {'Portugal': 0, 'Spain': 0, 'Morocco': 0, 'Iran': 0},
        'C': {'France': 0, 'Australia': 0, 'Peru': 0, 'Denmark': 0},
        'D': {'Argentina': 0, 'Iceland': 0, 'Croatia': 0, 'Nigeria': 0},
        'E': {'Brazil': 0, 'Switzerland': 0, 'Costa Rica': 0, 'Serbia': 0},
        'F': {'Germany': 0, 'Mexico': 0, 'Sweden': 0, 'Korea Republic': 0},
        'G': {'Belgium': 0, 'Panama': 0, 'Tunisia': 0, 'England': 0},
        'H': {'Poland': 0, 'Senegal': 0, 'Colombia': 0, 'Japan': 0}
    }

    wc_groups_dict = {}
    
    ### create team, group pair dictionary 
    for k, v in wc_groups.items():
        for i, j in v.items():
            wc_groups_dict[i] = k
    wc_groups_dict
    
    for i, row in wc_matches.iterrows():
        
        t1 = row['team1']
        t2 = row['team2']
            
        t1_g = wc_groups_dict[t1]
        t2_g = wc_groups_dict[t2]
        
        ### if the result is win
        ### it means team1 get 3 points 
        ### team1 get 0 point
        if row['result'] == 'win':
            wc_groups[t1_g][t1] += 3
        
        ### if the result is draw
        ### both teams get 1 point
        if row['result'] == 'draw':
            wc_groups[t1_g][t1] += 1
            wc_groups[t2_g][t2] += 1
            
        ### if the result is lose
        ### team1 get 0 point and team2 
        ### will get 3 points
        if row['result'] == 'lose':
            wc_groups[t2_g][t2] += 3
            
    return wc_groups
````

````python
wc_groups = simulate_group_stage(wc_matches)
wc_groups
````

__During the round of 16, the top 2 teams in group A will play against top 2 teams in group B. Teams in group C will play against teams in Group D .... Also, the first place team in each group will play against the second place team in the opponent group__

````python
def create_tmp_result(team1s, team2s):
    ### construct a dataframe to schedule all the matches
    ### in the round of 16.     
    tmp = pd.DataFrame(data = {'team1': team1s, 'team2': team2s})
    tmp = pd.merge(tmp, league_df, how='left', left_on='team1', right_on='name')
    tmp = pd.merge(tmp, league_df, how='left', left_on='team2', right_on='name')    
    tmp = pd.merge(tmp, recent_matches_df, how='left', left_on='fifa_code_x', right_on='team_name')
    tmp = pd.merge(tmp, recent_matches_df, how='left', left_on='fifa_code_y', right_on='team_name')
    tmp.rename(columns={'fifa_code_x': 'fifa_code_team1',
                   'fifa_code_y': 'fifa_code_team2',
                   'league_strength_x': 'league_strength_team1', 
                   'league_strength_y': 'league_strength_team2'}, inplace=True)
    
    tmp['team1Home'] = False
    tmp['team2Home'] = False

    tmp.loc[tmp.team1 == 'Russia', 'team1Home'] = True
    tmp.loc[tmp.team2 == 'Russia', 'team2Home'] = True
    
    tmp_predict = model.best_estimator_.predict(tmp[features])
    
    def text_result(predict):
        result = []
        
        for i in predict:
            if i > 0:
                result.append('win')
            else:
                result.append('lose')
        
        return result
    
    
    tmp_teams = []
    result = text_result(tmp_predict)
    
    for i, row in tmp.iterrows():
        if result[i] == 'win':
            tmp_teams.append(row['team1'])
        if result[i] == 'lose':
            tmp_teams.append(row['team2'])
    
    return tmp_teams
````

````python
def simulate_round_of_16(wc_groups):
    ### sort the dictionary just passed
    ### and the only keep the top 2 teams
    ### in each group
    for k, v in wc_groups.items():
        v = sorted(v.items(), key=lambda kv: kv[1], reverse=True)
        first = v[0][0]
        second = v[1][0]
        wc_groups[k] = [first, second]
    
    team1s, team2s = [], []
    
    for k, v in wc_groups.items():
        team1s.append(v[0])
        team2s.append(v[1])
    
    ### fix the order of team2s
    for i in range(0, len(team2s), 2):
        team2s[i], team2s[i+1] = team2s[i+1], team2s[i]
    
    return create_tmp_result(team1s, team2s)
````

````python
r16_result = simulate_round_of_16(wc_groups)
````

````python
def simulate_final_stage(r16_result):
    team1s, team2s = [], []
    
    for i in range(0, len(r16_result), 2):
        team1s.append(r16_result[i])
        team2s.append(r16_result[i+1])
        
    return team1s, team2s
````
